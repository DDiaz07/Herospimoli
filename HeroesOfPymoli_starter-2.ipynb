```python
SNNamesList = list(purchase_data['SN'].unique())
NumTimesPurchasedList = []
AgeList = []
for sn in SNNamesList:
    AgeList.append(purchase_data['Age'].loc[purchase_data['SN']==sn].max())
    NumTimesPurchasedList.append(purchase_data['SN'].loc[purchase_data['SN']==sn].count())
```

```python
newdfreallynewthistime = purchase_data.groupby('SN').sum()
newdfreallynewthistime['Number of Purchases Made']=NumTimesPurchasedList
newdfreallynewthistime
```
